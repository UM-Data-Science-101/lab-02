#### 4.b

Copilot can also explain code. Here is some prewritten Python code. Highlight the code, right click, and ask Copilot to explain what is going on by clicking on the sparkle icon.

````python
def power(base, exponent):
    result = 1
    while exponent:
        if exponent & 1:
            result *= base
        base *= base
        exponent >>= 1
    return result
````
